```python
import pandas as pd
import numpy as np
```

```python
titanic_dataset = pd.read_csv('train.csv')
titanic_dataset
```

```python
titanic_dataset.isna().sum()
```

```python
del titanic_dataset['Cabin']
titanic_dataset
```

```python
titanic_dataset.isna().sum()
```

```python
del titanic_dataset['Ticket']
del titanic_dataset['Name']
```

```python
titanic_dataset
```

```python
titanic_dataset['Embarked'].value_counts()
```

```python
titanic_dataset['Embarked'].fillna(titanic_dataset['Embarked'].mode()[0],inplace=True)
```

```python
titanic_dataset.isna().sum()
```

```python
titanic_dataset['Age'].mean()
```

```python
titanic_dataset['Age'].median()
```

```python
ages = titanic_dataset.groupby('Pclass', as_index=False)['Age'].median().Age.tolist()
for i in range(len(ages)):
	titanic_dataset.loc[titanic_dataset.Age.isna() & (titanic_dataset.Pclass == i+1), "Age"] = ages[i]
ages
```

```python
titanic_dataset['Age'].fillna(titanic_dataset['Age'].mean(),inplace=True)
```

```python
titanic_dataset.isna().sum()
```

```python
dummies = []
cols = ['Pclass', 'Sex', 'Embarked']
for col in cols:
	dummies.append(pd.get_dummies(titanic_dataset[col]))
all_dummies = pd.concat(dummies, axis=1)
all_dummies
```

```python
titanic_dataset = pd.concat((titanic_dataset,all_dummies), axis=1)
titanic_dataset = titanic_dataset.drop(['Pclass', 'Sex', 'Embarked'], axis=1)
```
